### Imports & Set up file paths

```python
import pandas as pd
import os

pd.set_option('display.max_columns', 60)

srcpath = os.path.split(os.path.abspath(''))[0]
rootpath = os.path.split(srcpath)[0]
datapath = os.path.join(srcpath, 'data')
rawpath = os.path.join(datapath, 'raw')
respath = os.path.join(rawpath, 'reservations_rec_gov/')
tabpath = os.path.join(rawpath, 'tables_rec_gov/')
```

### Generate lists of files in the raw/reservations directory

```python
list_reservations = []
for root, dirs, file in os.walk(respath):
    list_reservations.extend(file)
x = list_reservations.pop(0)
x == '.DS_Store'
```

```python
for year in sorted(list_reservations):
    df = pd.read_csv(respath + year, encoding='latin-1')
    print(f'{year}')
    print([col for col in df.columns])
    #display(df.head(low_memory=False))
    display(df.head())
    display(df.info(null_counts=True))
    print('\n\n\n')
```
