### Imports & Set up file paths

```python
import pandas as pd
import os

srcpath = os.path.split(os.path.abspath(''))[0]
rootpath = os.path.split(srcpath)[0]
datapath = os.path.join(srcpath, 'data')
rawpath = os.path.join(datapath, 'raw')
respath = os.path.join(rawpath, 'reservations_rec_gov/')
tabpath = os.path.join(rawpath, 'tables_rec_gov/')
```

### Generate lists of files in each raw/tables directory

```python
list_tables = []
for root, dirs, file in os.walk(tabpath):
    list_tables.extend(file)
```

### Display attributes for all tables

```python
for table in sorted(list_tables):
    df = pd.read_csv(tabpath + table, low_memory=False)
    print(f'{table}')
    print([col for col in df.columns])
    display(df.head())
    display(df.info())
    print('\n\n')
```
